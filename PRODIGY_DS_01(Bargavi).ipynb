```python
import pandas as pd

# Load the dataset
data = pd.read_csv('C:\\Users\\Bargavi\\Desktop\\population.csv')

# Ensure '2022' column is numeric
data['2022'] = pd.to_numeric(data['2022'], errors='coerce')

# Filter out known non-country entries
non_countries = [
    'World', 'High income', 'Low income', 'Upper middle income', 'Lower middle income',
    'Middle income', 'East Asia & Pacific', 'Europe & Central Asia', 'Latin America & Caribbean',
    'Sub-Saharan Africa', 'South Asia', 'North America', 'Arab World', 'IDA only',
    'IDA & IBRD total', 'IBRD only', 'Euro area', 'Post-demographic dividend',
    'Pre-demographic dividend', 'Heavily indebted poor countries (HIPC)'
]
country_data = data[~data['Country Name'].isin(non_countries)]

# Calculate total population for 2022
total_population_2022 = country_data['2022'].sum()

# Print result
print(f"Total population (across all countries) in 2022: {total_population_2022:,.0f}")

# Calculate total population for 2001
total_population_2022 = country_data['2001'].sum()

# Print result
print(f"Total population (across all countries) in 2001: {total_population_2022:,.0f}")
```

```python
# Filter data for total population
total_population_data = data[data['Series Code'] == 'SP.POP.TOTL']

# Sort data based on the total population for 2022
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)

# Get the top ten countries with the highest total population for 2022
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population in 2022\n")
print(total_top_ten_countries[['Country Name']] )


# Sort data based on the total population for 2001
total_population_sorted = total_population_data.sort_values(by="2001", ascending=False)

# Get the top ten countries with the highest total population for 2001
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population in 2001\n")
print(total_top_ten_countries[['Country Name']] )
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the total_top_ten_countries DataFrame first
# Example of how to create this DataFrame (you should replace with your actual data)
data = {
    'Country Name': ['China', 'India', 'United States', 'Indonesia', 'Pakistan', 
                     'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico'],
    '2001': [1273000000, 1059000000, 285000000, 216000000, 144000000,
             176000000, 126000000, 132000000, 146000000, 103000000],
    '2022': [1412000000, 1380000000, 332000000, 275000000, 235000000,
             215000000, 214000000, 171000000, 144000000, 130000000]
}
total_top_ten_countries = pd.DataFrame(data)

# Create the bar plot for 2022
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="Country Name", y="2022", data=total_top_ten_countries, palette="viridis")
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Country",fontsize=4)
plt.ylabel("Total Population",fontsize=4)
plt.xticks(rotation=90)
plt.show()

# Create the bar plot for 2001
plt.figure(figsize=(10, 6))
plt.subplot(2,2,2)
sns.barplot(x="Country Name", y="2001", data=total_top_ten_countries, palette="plasma")
plt.title("Top Ten Countries with Total Population (2001)",fontsize=10)
plt.xlabel("Country",fontsize=4)
plt.ylabel("Total Population",fontsize=4)
plt.xticks(rotation=90)
plt.show()
```
